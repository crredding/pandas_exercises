```python
def capitalizer(x):
    return ''.join([letter.capitalize() for letter in x])
        
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(x):
    '''
    Takes in an individual's age and outputs True if they are above age 16.
    
    Input:
    ------
    age: int - individual's age
    
    Output:
    -------
    legal_drinker: bool - True if legal drinker
    '''
    return x > 16
```

```python
df_sliced['legal_drinker'] = df_sliced.apply(lambda row: majority(row['age']), axis=1)
```
